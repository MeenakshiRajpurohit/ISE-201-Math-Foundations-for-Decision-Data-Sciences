## Question 1
Provide a reasonable description of the sample space for each of the random experiments. There can be more than one acceptable interpretation of each experiment. Describe any assumptions you make.



### Random Experiment:
Each of four transmitted bits is classified as either **in error** or **not in error**.



### Answer:
A reasonable description of the **sample space** for the random experiment is the set of all possible outcomes for classifying four transmitted bits as either **in error (E)** or **not in error (N)**.



### Sample Space Description

Let’s denote the classification of a single bit as:
- **E** for "in error"
- **N** for "not in error"

Since there are four independent bits, the outcome of the experiment can be represented as an ordered sequence (or 4-tuple) of the classifications for the four bits.



### Set of Outcomes

The sample space **S** is the set of all sequences of length four where each element is either E or N:

S = {(b1, b2, b3, b4) | bi ∈ {E, N} for i = 1, 2, 3, 4}



### Listing the Outcomes

Since there are 2 possible classifications for each of the 4 bits, the total number of outcomes is:

2^4 = 16

The 16 possible outcomes in the sample space S are:

(N, N, N, N)  (No errors)  
(E, N, N, N)  
(N, E, N, N)  
(N, N, E, N)  
(N, N, N, E)  (One error)  

(E, E, N, N)  
(E, N, E, N)  
(E, N, N, E)  
(N, E, E, N)  
(N, E, N, E)  
(N, N, E, E)  (Two errors)  

(E, E, E, N)  
(E, E, N, E)  
(E, N, E, E)  
(N, E, E, E)  (Three errors)  

(E, E, E, E)  (Four errors)



### Alternative Interpretation

A more compact sample space may focus on the **number of bits in error**, instead of their positions. In this case:

Scount = {0, 1, 2, 3, 4}



### Assumptions Made

- Each of the four bits is **distinguishable** (e.g., ordered by transmission time or position).
- Therefore, (E, N, N, N) is considered **different** from (N, E, N, N).

This is a standard assumption when describing the full outcome space for a sequence of events.

### Question 5:

The diameter of a particle of contamination (in
micrometers) is modeled with the probability density function
f(x) = 2∕x3 for x > 1.

Determine the following:

a. P(X < 2)

b. P(X > 5)

c. P(4 < X < 8)

 d. P(X < 4 or X > 8)


e. x such that P(X < x) = 0.95

**Answer:**

##### Continuous Random Variable Probability Calculations

This problem involves calculating probabilities and a quantile for a continuous random variable $X$ with the probability density function (PDF):

$$f(x) = \frac{2}{x^3} \quad \text{for } x > 1$$

$$f(x) = 0 \quad \text{otherwise}$$

For a continuous random variable, the probability $P(a < X < b)$ is found by integrating the PDF over the interval:

$$P(a < X < b) = \int_a^b f(x) dx$$

####Cumulative Distribution Function (CDF)

First, we find the Cumulative Distribution Function (CDF), $F(x)$, which is needed for most parts of the problem:

$$F(x) = P(X \leq x) = \int_{1}^{x} \frac{2}{u^3} du \quad \text{for } x > 1$$

$$\int 2u^{-3} du = 2 \cdot \frac{u^{-2}}{-2} = -\frac{1}{u^2}$$

$$F(x) = \left. -\frac{1}{u^2} \right|_1^x = \left( -\frac{1}{x^2} \right) - \left( -\frac{1}{1^2} \right) = 1 - \frac{1}{x^2}$$

So, the CDF is:

$$F(x) = 1 - \frac{1}{x^2} \quad \text{for } x \geq 1$$

$$F(x) = 0 \quad \text{for } x < 1$$



##### a. $P(X < 2)$

This is the CDF evaluated at $x=2$: $P(X < 2) = F(2)$.

$$P(X < 2) = 1 - \frac{1}{2^2} = 1 - \frac{1}{4} = \mathbf{0.75}$$




##### b. $P(X > 5)$

This is the complement of $P(X \leq 5)$: $P(X > 5) = 1 - F(5)$.

$$P(X > 5) = 1 - \left( 1 - \frac{1}{5^2} \right) = \frac{1}{25} = \mathbf{0.04}$$



##### c. $P(4 < X < 8)$

We calculate this using the CDF: $P(4 < X < 8) = F(8) - F(4)$.

$$P(4 < X < 8) = \left( 1 - \frac{1}{8^2} \right) - \left( 1 - \frac{1}{4^2} \right)$$

$$P(4 < X < 8) = \frac{1}{4^2} - \frac{1}{8^2} = \frac{1}{16} - \frac{1}{64}$$

$$P(4 < X < 8) = \frac{4}{64} - \frac{1}{64} = \frac{3}{64} \approx \mathbf{0.046875}$$




##### d. $P(X < 4 \text{ or } X > 8)$

Since the events $X < 4$ and $X > 8$ are mutually exclusive, we can write:

$$P(X < 4 \text{ or } X > 8) = P(X < 4) + P(X > 8)$$

We use the CDF for both terms:

$$P(X < 4) = F(4) = 1 - \frac{1}{4^2} = 1 - \frac{1}{16} = \frac{15}{16}$$

$$P(X > 8) = 1 - F(8) = 1 - \left( 1 - \frac{1}{8^2} \right) = \frac{1}{64}$$

$$P(X < 4 \text{ or } X > 8) = \frac{15}{16} + \frac{1}{64} = \frac{60}{64} + \frac{1}{64} = \frac{61}{64} \approx \mathbf{0.953125}$$

**Alternatively**, this is the complement of the event $4 < X < 8$:

$$P(X < 4 \text{ or } X > 8) = 1 - P(4 < X < 8) = 1 - \frac{3}{64} = \frac{64 - 3}{64} = \frac{61}{64}$$



##### e. $x$ such that $P(X < x) = 0.95$

We need to solve $F(x) = 0.95$ for $x$:

$$F(x) = 1 - \frac{1}{x^2} = 0.95$$

$$1 - 0.95 = \frac{1}{x^2}$$

$$0.05 = \frac{1}{x^2}$$

$$x^2 = \frac{1}{0.05} = 20$$

$$x = \sqrt{20} = \sqrt{4 \cdot 5} = 2\sqrt{5}$$

$$x \approx \mathbf{4.472}$$

###Question 7

The systolic and diastolic blood pressure values (mm Hg)
are the pressures when the heart muscle contracts and relaxes
(denoted as Y and X, respectively). Over a collection of individuals, the distribution of diastolic pressure is normal with mean 73
and standard deviation 8. The systolic pressure is conditionally normally distributed with mean 1.6x when X = x and standard
deviation of 10. Determine the following:

a. Conditional probability density function of Y given

b. P(Y < 115 | X = 73)

c. E(Y | X = 73)

d. Recognize the distribution fXY (x, y) and identify the mean

**Answer**

### Bivariate Normal Distribution: Blood Pressure Analysis

This problem deals with a bivariate normal distribution, where the joint distribution of diastolic pressure ($X$) and systolic pressure ($Y$) is normal. We can determine the requested values based on the properties of normal and conditional normal distributions.

#### Given Information

* **Diastolic Pressure ($X$)**: $X \sim N(\mu_X, \sigma_X^2)$
   * Mean: $\mu_X = 73$ mm Hg
   * Standard Deviation: $\sigma_X = 8$
* **Systolic Pressure ($Y$)**: Conditionally distributed as $Y | X=x \sim N(\mu_{Y|x}, \sigma_{Y|x}^2)$
   * Conditional Mean: $E(Y | X=x) = 1.6x$
   * Conditional Standard Deviation: $\sigma_{Y|x} = 10$



#### a. Conditional Probability Density Function of $Y$ given $X = 73$

The conditional distribution of $Y$ given $X=x$ is Normal. Therefore, the probability density function (PDF) for $Y | X=73$ is the standard Normal PDF:

$$f_{Y|X}(y | x) = \frac{1}{\sqrt{2\pi \sigma_{Y|x}^2}} \exp\left\{-\frac{1}{2}\left(\frac{y - \mu_{Y|x}}{\sigma_{Y|x}}\right)^2\right\}$$

##### 1. Calculate the Conditional Mean:

$$E(Y | X=73) = \mu_{Y|73} = 1.6x = 1.6(73) = \mathbf{116.8}$$

##### 2. State the Conditional Variance/Std Dev:

$$\sigma_{Y|73} = 10 \quad \text{and} \quad \sigma_{Y|73}^2 = 100$$

##### 3. The Conditional PDF:

Substitute the values into the Normal PDF formula:

$$f_{Y|X}(y | 73) = \frac{1}{\sqrt{2\pi (100)}} \exp\left\{-\frac{1}{2}\left(\frac{y - 116.8}{10}\right)^2\right\}$$



#### b. $P(Y < 115 | X = 73)$

This is a probability calculation using the conditional distribution $Y | X=73 \sim N(116.8, 10^2)$. We first standardize the value $Y=115$ using the conditional mean and standard deviation:

####3 1. Calculate the z-score:

$$z = \frac{Y - \mu_{Y|73}}{\sigma_{Y|73}} = \frac{115 - 116.8}{10} = \frac{-1.8}{10} = \mathbf{-0.18}$$

##### 2. Find the Probability:

$$P(Y < 115 | X = 73) = P(Z < -0.18)$$

Using the standard normal table or calculator for $P(Z < -0.18)$:

$$P(Z < -0.18) \approx \mathbf{0.4286}$$



#### c. $E(Y | X = 73)$

The conditional expected value, $E(Y | X=73)$, is simply the conditional mean $\mu_{Y|73}$, which we calculated in part (a):

$$E(Y | X = 73) = 1.6(73) = \mathbf{116.8}$$



#### d. Recognize the distribution $f_{XY}(x, y)$ and identify the mean and variance of $Y$

##### Distribution Recognition

The joint probability density function $f_{XY}(x, y)$ can be written as:

$$f_{XY}(x, y) = f_{Y|X}(y | x) \cdot f_X(x)$$

Since $X$ is Normal and the conditional distribution $Y|X$ is Normal with a linear conditional mean ($1.6x$) and a constant conditional variance ($\sigma_{Y|x}^2 = 100$), the joint distribution $f_{XY}(x, y)$ is a **Bivariate Normal Distribution**.

#### Mean and Variance of $Y$

To find the mean $E(Y)$ and variance $\text{Var}(Y)$, we use the formulas derived from the properties of conditional distributions:

##### 1. Mean of $Y$: $E(Y) = E[E(Y | X)]$

$$E(Y) = E[1.6X]$$



$$E(Y) = 1.6 \cdot E(X) = 1.6 \cdot \mu_X$$

Substitute $\mu_X = 73$:

$$E(Y) = 1.6(73) = \mathbf{116.8}$$

###### 2. Variance of $Y$: $\text{Var}(Y) = E[\text{Var}(Y | X)] + \text{Var}[E(Y | X)]$

**Term 1:** $E[\text{Var}(Y | X)]$

Since $\text{Var}(Y | X=x) = 100$ (constant), the expected value is simply 100.

$$E[\text{Var}(Y | X)] = 100$$

**Term 2:** $\text{Var}[E(Y | X)]$

$$\text{Var}[E(Y | X)] = \text{Var}[1.6X]$$

Using $\text{Var}(cX) = c^2 \cdot \text{Var}(X)$:

$$\text{Var}[1.6X] = (1.6)^2 \cdot \text{Var}(X) = 2.56 \cdot \sigma_X^2$$

Substitute $\sigma_X = 8$:

$$\text{Var}[1.6X] = 2.56 \cdot (8)^2 = 2.56 \cdot 64 = 163.84$$

**Total Variance:**

$$\text{Var}(Y) = 100 + 163.84 = \mathbf{263.84}$$

##### Summary for $Y$:

* **Mean of Y:** $E(Y) = \mathbf{116.8}$ mm Hg
* **Variance of Y:** $\text{Var}(Y) = \mathbf{263.84}$ $\text{(mm Hg)}^2$

###Question 8:

Determine the covariance and correlation
for the following joint probability distribution:

x             1 1 2 4

y             3 4 5 6

fXY (x, y)    1/8 1/4 1/2 1/8


**Answer**
#### Covariance and Correlation Calculation





| $x$ | $y$   | f(x,y) |
|-----|------ |--------|
| 1   | 3     | 1/8    |
| 1   | 4     | 1/4    |
| 2   | 5     | 1/2    |
| 4   | 6     | 1/8    |





##### 1. Compute $E[X]$

$$E[X] = 1\cdot\frac{1}{8} + 1\cdot\frac{1}{4} + 2\cdot\frac{1}{2} + 4\cdot\frac{1}{8}$$

$$= \frac{1}{8} + \frac{1}{4} + 1 + \frac{4}{8} = \frac{1}{8} + \frac{2}{8} + \frac{8}{8} + \frac{4}{8} = \frac{15}{8} = 1.875$$




##### 2. Compute $E[Y]$

$$E[Y] = 3\cdot\frac{1}{8} + 4\cdot\frac{1}{4} + 5\cdot\frac{1}{2} + 6\cdot\frac{1}{8}$$

$$= \frac{3}{8} + 1 + \frac{5}{2} + \frac{6}{8} = \frac{3}{8} + \frac{4}{4} + \frac{10}{4} + \frac{6}{8} = \frac{3}{8} + \frac{8}{8} + \frac{20}{8} + \frac{6}{8} = \frac{37}{8} = 4.625$$



##### 3. Compute $E[XY]$

$$E[XY] = (1\cdot 3)\cdot\frac{1}{8} + (1\cdot 4)\cdot\frac{1}{4} + (2\cdot 5)\cdot\frac{1}{2} + (4\cdot 6)\cdot\frac{1}{8}$$

$$= 3\cdot\frac{1}{8} + 4\cdot\frac{1}{4} + 10\cdot\frac{1}{2} + 24\cdot\frac{1}{8}$$

$$= \frac{3}{8} + 1 + 5 + \frac{24}{8} = \frac{3}{8} + \frac{8}{8} + \frac{40}{8} + \frac{24}{8} = \frac{75}{8} = 9.375$$



##### 4. Covariance

$$\text{Cov}(X,Y) = E[XY] - E[X]E[Y] = \frac{75}{8} - \frac{15}{8}\cdot\frac{37}{8}$$



$$\frac{15}{8} \cdot \frac{37}{8} = \frac{555}{64}$$

Convert $E[XY]$ to /64:

$$\frac{75}{8} = \frac{600}{64}$$

$$\text{Cov}(X,Y) = \frac{600}{64} - \frac{555}{64} = \frac{45}{64} = 0.703125$$



##### 5. Variances of $X$ and $Y$

##### Var(X)

$$E[X^2] = 1^2\cdot\frac{1}{8} + 1^2\cdot\frac{1}{4} + 2^2\cdot\frac{1}{2} + 4^2\cdot\frac{1}{8}$$

$$= \frac{1}{8} + \frac{1}{4} + 4\cdot\frac{1}{2} + 16\cdot\frac{1}{8} = \frac{1}{8} + \frac{2}{8} + 2 + 2 = \frac{1}{8} + \frac{2}{8} + \frac{16}{8} + \frac{16}{8} = \frac{35}{8}$$

$$\text{Var}(X) = \frac{35}{8} - \left(\frac{15}{8}\right)^2 = \frac{35}{8} - \frac{225}{64} = \frac{280}{64} - \frac{225}{64} = \frac{55}{64}$$

##### Var(Y)

$$E[Y^2] = 3^2\cdot\frac{1}{8} + 4^2\cdot\frac{1}{4} + 5^2\cdot\frac{1}{2} + 6^2\cdot\frac{1}{8}$$

$$= 9\cdot\frac{1}{8} + 16\cdot\frac{1}{4} + 25\cdot\frac{1}{2} + 36\cdot\frac{1}{8} = \frac{9}{8} + 4 + 12.5 + \frac{36}{8}$$

$$= \frac{9}{8} + \frac{32}{8} + \frac{100}{8} + \frac{36}{8} = \frac{177}{8}$$

$$\text{Var}(Y) = \frac{177}{8} - \left(\frac{37}{8}\right)^2 = \frac{177}{8} - \frac{1369}{64} = \frac{1416}{64} - \frac{1369}{64} = \frac{47}{64}$$



##### 6. Correlation

$$\rho = \frac{\text{Cov}(X,Y)}{\sqrt{\text{Var}(X)\text{Var}(Y)}} = \frac{\frac{45}{64}}{\sqrt{\frac{55}{64}\cdot\frac{47}{64}}} = \frac{45/64}{\frac{\sqrt{55\cdot47}}{64}} = \frac{45}{\sqrt{2585}} \approx \frac{45}{50.84} \approx 0.885$$



| Quantity | Value |
|----------|-------|
| $\text{Cov}(X,Y)$ | $\frac{45}{64} \approx 0.7031$ |
| $\text{Corr}(X,Y)$ | $\approx 0.885$ |

This is a **strong positive correlation**, meaning larger $X$ values are strongly associated with larger $Y$ values in this distribution.
