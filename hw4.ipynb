**Question 9.2.6** A manufacturer produces crankshafts for an automobile engine. The crankshafts wear after 100,000 miles (0.0001 inch) is of interest because it is likely to have an impact on warranty claims. A random sample of n = 15 shafts is tested and x = 2.78. It is known that σ = 0.9 and that wear is normally distributed.

a. Test H0: μ = 3 versus H1: μ ≠ 3 using α = 0.05.

b. What is the power of this test if μ = 3.25?

c. What sample size would be required to detect a true mean of 3.75 if we wanted the power to be at least 0.9?



Answer:


### Hypothesis Testing and Power Analysis

#### Given:  

* Population Standard Deviation (σ): 0.9
* Sample Size (n): 15
* Sample Mean (x̄): 2.78
* Significance Level (α): 0.05



#### Part a: Test H₀: μ = 3 versus H₁: μ ≠ 3 using α = 0.05

##### 1. Hypotheses

* **Null Hypothesis (H₀):** μ = 3
* **Alternative Hypothesis (H₁):** μ ≠ 3 (This is a two-tailed test)

#### 2. Test Statistic (Z-score)

We use the formula for the Z-statistic:

$$Z = \frac{\bar{x} - \mu_0}{\sigma / \sqrt{n}}$$

$$Z = \frac{2.78 - 3.0}{0.9 / \sqrt{15}} = \frac{-0.22}{0.9 / 3.873} \approx \frac{-0.22}{0.2324}$$

$$\mathbf{Z \approx -0.947}$$

#### 3. Critical Values (α = 0.05)

For a two-tailed test with α = 0.05, we look up Z_(α/2) = Z_(0.025).

* Critical Z-values are **±1.96**
* The rejection region is Z < -1.96 or Z > 1.96

#### 4. Conclusion

Since the calculated Z-score (Z = -0.947) is greater than the lower critical value (-1.96) and less than the upper critical value (1.96), it falls within the acceptance region.

* Do not reject H₀
* There is insufficient statistical evidence at the 5% significance level to conclude that the mean crankshaft wear is different from 3.



#### Part b: What is the power of this test if μ = 3.25?

Power (Power = 1 - β) is the probability of correctly rejecting the null hypothesis when the true mean (μ₁) is 3.25.

#### 1. Determine Critical Sample Means (x̄_critical)

We first find the values of x̄ that define the rejection region under H₀: μ = 3:

$$\bar{x}_{critical} = \mu_0 \pm Z_{\alpha/2} \left(\frac{\sigma}{\sqrt{n}}\right)$$

$$\bar{x}_{critical} = 3 \pm 1.96 \times 0.2324 \approx 3 \pm 0.4555$$

* Lower critical mean (x̄_L): 2.5445
* Upper critical mean (x̄_U): 3.4555
* Rejection region: x̄ < 2.5445 or x̄ > 3.4555

#### 2. Calculate β (Type II Error) under μ₁ = 3.25

β is the probability of failing to reject H₀ when μ₁ = 3.25. This occurs when 2.5445 < x̄ < 3.4555. We convert these boundaries to Z-scores using the true mean μ₁ = 3.25:

$$Z_L = \frac{2.5445 - 3.25}{0.2324} \approx -3.036$$

$$Z_U = \frac{3.4555 - 3.25}{0.2324} \approx 0.884$$

$$\beta = P(-3.036 < Z < 0.884) = P(Z < 0.884) - P(Z < -3.036)$$

$$\beta \approx 0.8115 - 0.0012 = 0.8103$$

#### 3. Calculate Power

$$\text{Power} = 1 - \beta = 1 - 0.8103$$

$$\mathbf{\text{Power} \approx 0.1897}$$



#### Part c: What sample size would be required to detect a true mean of 3.75 if we wanted the power to be at least 0.9?

We need to find n such that the power is 0.9, meaning β = 1 - 0.9 = 0.1.

#### 1. Define Parameters

* H₀: μ₀ = 3
* μ₁ = 3.75 (True mean to detect)
* |μ₁ - μ₀| = 0.75
* σ = 0.9
* Z_(α/2) for α=0.05 (two-tailed): 1.96
* Z_β for β=0.10: 1.28

#### 2. Sample Size Formula (Two-Tailed Test)




 $$n = \left[\frac{(Z_{\alpha/2} + Z_{\beta})\sigma}{\mu_1 - \mu_0}\right]^2$$



$$n = \left[\frac{(1.96 + 1.28) \times 0.9}{3.75 - 3.0}\right]^2$$

$$n = \left[\frac{3.24 \times 0.9}{0.75}\right]^2 = \left[\frac{2.916}{0.75}\right]^2$$

$$n = (3.888)^2 \approx 15.1165$$



Since the sample size must be a whole number, and we require the power to be at least 0.9, we must round up to the next integer.

$$\mathbf{n = 16}$$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
from math import ceil

# --- Input Parameters ---
sigma = 0.9          # Population standard deviation
n = 15               # Sample size
x_bar = 2.78         # Sample mean
mu_0 = 3.0           # Null hypothesis mean
alpha = 0.05         # Significance level
mu_1 = 3.25          # Alternative mean for Part b
mu_2 = 3.75          # Alternative mean for Part c
power_target = 0.9   # Target power for Part c

# --- Calculations ---
SE = sigma / np.sqrt(n) # Standard Error

# --- Part A: Hypothesis Test (Two-Tailed) ---
print("--- Part A: Hypothesis Test (μ₀ = 3.0) ---")

# Z-test statistic
Z_calc = (x_bar - mu_0) / SE
print(f"Calculated Z-statistic: {Z_calc:.4f}")

# Critical Z-values for two-tailed test
Z_critical_upper = norm.ppf(1 - alpha / 2)
Z_critical_lower = norm.ppf(alpha / 2)
print(f"Critical Z-values (α=0.05): ±{Z_critical_upper:.3f}")
print(f"Decision: {'Do Not Reject H₀' if abs(Z_calc) < Z_critical_upper else 'Reject H₀'}\n")


# --- Part B: Power Calculation (μ₁ = 3.25) ---
print("--- Part B: Power Calculation (μ₁ = 3.25) ---")

# 1. Critical sample means (x̄_critical) under H₀
x_critical_L = mu_0 + Z_critical_lower * SE
x_critical_U = mu_0 + Z_critical_upper * SE
print(f"Lower Critical Mean (x̄_L): {x_critical_L:.4f}")
print(f"Upper Critical Mean (x̄_U): {x_critical_U:.4f}")

# 2. Calculate beta (Type II Error) under H₁
# Convert critical means to Z-scores under H₁ (μ₁=3.25)
Z_L_under_H1 = (x_critical_L - mu_1) / SE
Z_U_under_H1 = (x_critical_U - mu_1) / SE

# Beta is the area between Z_L and Z_U under the H₁ curve
beta = norm.cdf(Z_U_under_H1) - norm.cdf(Z_L_under_H1)
power = 1 - beta
print(f"Beta (β): {beta:.4f}")
print(f"Power (1 - β): {power:.4f}\n")


# --- Part C: Sample Size Calculation (μ₂ = 3.75, Power = 0.9) ---
print("--- Part C: Sample Size for Target Power (μ₂ = 3.75) ---")

# Z-values for target alpha and beta
Z_alpha_half = norm.ppf(1 - alpha / 2)  # 1.96
Z_beta = norm.ppf(power_target)        # For Power=0.9, Z_beta = norm.ppf(0.9) ≈ 1.28

# Sample size formula for two-tailed test
delta_mu = abs(mu_2 - mu_0)
n_required = ((Z_alpha_half + Z_beta) * sigma / delta_mu)**2

# Round up to the next whole number
n_final = ceil(n_required)

print(f"Required theoretical sample size: {n_required:.4f}")
print(f"Final required sample size (rounded up): {n_final}\n")


# --- Visualization Setup ---
x = np.linspace(mu_0 - 3.5 * SE, mu_1 + 3.5 * SE, 500)
# Use a wider range for the second plot for better context
x_wide = np.linspace(2.0, 4.0, 500)
y0 = norm.pdf(x_wide, mu_0, SE)
y1 = norm.pdf(x_wide, mu_1, SE)


plt.style.use('seaborn-v0_8-darkgrid')
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 10))
fig.suptitle('Crankshaft Wear Statistical Analysis', fontsize=16, fontweight='bold')

# ------------------------------------------
# Plot 1: Hypothesis Test (Part A)
# ------------------------------------------
ax1.plot(x_wide, y0, label=r'$H_0: \mu = 3.0$', color='#1f77b4', linewidth=2)
ax1.fill_between(x_wide, y0, where=(x_wide < x_critical_L) | (x_wide > x_critical_U),
                 color='red', alpha=0.2, hatch='/', label=r'Rejection Region ($\alpha$)')

# Vertical lines for critical values
ax1.axvline(x_critical_L, color='red', linestyle='--', linewidth=1)
ax1.axvline(x_critical_U, color='red', linestyle='--', linewidth=1)

# Plot sample mean (x̄ = 2.78)
y_xbar = norm.pdf(x_bar, mu_0, SE)
ax1.axvline(x_bar, color='black', linestyle='-', linewidth=2, label=r'Sample Mean ($\bar{x}=2.78$)')
ax1.text(x_bar, y_xbar * 0.95, r'$\bar{x}=2.78$', ha='center', color='black', fontweight='bold')

# Labels and Titles for Plot 1
ax1.set_title(r'Part A: Hypothesis Test for $H_0: \mu = 3.0$', fontsize=14)
ax1.set_xlabel('Mean Wear ($\mu$) in inches $\times 10^{-4}$', fontsize=12)
ax1.set_ylabel('Probability Density', fontsize=12)
ax1.legend(loc='upper right')
ax1.grid(True, which='both', linestyle=':', linewidth=0.5)

# ------------------------------------------
# Plot 2: Power Analysis (Part B)
# ------------------------------------------
ax2.plot(x_wide, y0, label=r'$H_0: \mu = 3.0$', color='#1f77b4', linestyle='--')
ax2.plot(x_wide, y1, label=r'$H_1: \mu = 3.25$', color='#2ca02c', linewidth=2)

# Shade Beta (Type II Error - Failure to reject H₀ when H₁ is true)
x_beta = np.linspace(x_critical_L, x_critical_U, 100)
y1_beta = norm.pdf(x_beta, mu_1, SE)
ax2.fill_between(x_beta, y1_beta, color='orange', alpha=0.5, label=r'Type II Error ($\beta \approx 0.8103$)')

# Shade Power (Correctly Rejecting H₀ when H₁ is true)
x_power_U = np.linspace(x_critical_U, mu_1 + 3.5 * SE, 100)
y1_power_U = norm.pdf(x_power_U, mu_1, SE)
ax2.fill_between(x_power_U, y1_power_U, color='green', alpha=0.5, label=r'Power ($1 - \beta \approx 0.1897$)')

# Vertical lines for critical values
ax2.axvline(x_critical_L, color='red', linestyle=':', linewidth=1)
ax2.axvline(x_critical_U, color='red', linestyle=':', linewidth=1)
ax2.text(x_critical_U, norm.pdf(x_critical_U, mu_1, SE) * 1.05, r'$\bar{x}_U$', ha='center', color='red', fontsize=10)

# Labels and Titles for Plot 2
ax2.set_title(r'Part B: Power Calculation (True Mean $\mu_1 = 3.25$)', fontsize=14)
ax2.set_xlabel('Mean Wear ($\mu$) in inches $\times 10^{-4}$', fontsize=12)
ax2.set_ylabel('Probability Density', fontsize=12)
ax2.legend(loc='upper left')
ax2.grid(True, which='both', linestyle=':', linewidth=0.5)

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# ============================
# GIVEN PARAMETERS
# ============================

sigma = 0.9        # population standard deviation
n = 15             # sample size
x_bar = 2.78       # sample mean
mu0 = 3.0          # null hypothesis mean
alpha = 0.05       # significance level

# For power calculations
mu1_partb = 3.25   # true mean for part (b)
mu1_partc = 3.75   # true mean to detect for part (c)
power_target = 0.90

# Standard error
se = sigma / np.sqrt(n)

# ============================
# PART (a): HYPOTHESIS TEST
# ============================

# Test statistic
z_stat = (x_bar - mu0) / se

# Critical z-values for two-sided test
z_crit = norm.ppf(1 - alpha/2)  # ≈ 1.96
lower_crit = mu0 - z_crit * se
upper_crit = mu0 + z_crit * se

# p-value for two-sided test
p_value = 2 * (1 - norm.cdf(abs(z_stat)))

# Decision
reject_H0 = abs(z_stat) > z_crit

print("=== Part (a): Hypothesis Test ===")
print(f"Z-statistic: {z_stat:.3f}")
print(f"Critical z-values: ±{z_crit:.3f}")
print(f"Critical region for x̄: x̄ < {lower_crit:.4f} or x̄ > {upper_crit:.4f}")
print(f"p-value (two-sided): {p_value:.4f}")
print("Decision:", "Reject H0" if reject_H0 else "Fail to reject H0")
print()

# Visualization of Part (a)
x_values = np.linspace(mu0 - 4*se, mu0 + 4*se, 400)
pdf_H0 = norm.pdf(x_values, loc=mu0, scale=se)

plt.figure(figsize=(8, 5))
plt.plot(x_values, pdf_H0, label="Sampling distribution under H₀: N(3, σ/√n)")

# Shade rejection regions
x_left = np.linspace(mu0 - 4*se, lower_crit, 200)
x_right = np.linspace(upper_crit, mu0 + 4*se, 200)
plt.fill_between(x_left, norm.pdf(x_left, mu0, se), alpha=0.3, label="Rejection region (left)")
plt.fill_between(x_right, norm.pdf(x_right, mu0, se), alpha=0.3, label="Rejection region (right)")

# Mark sample mean
plt.axvline(x_bar, linestyle="--", linewidth=2, label=f"Sample mean x̄ = {x_bar:.2f}")

plt.axvline(lower_crit, linestyle=":", linewidth=1.5, label=f"Lower critical mean = {lower_crit:.2f}")
plt.axvline(upper_crit, linestyle=":", linewidth=1.5, label=f"Upper critical mean = {upper_crit:.2f}")

plt.title("Part (a): Hypothesis Test Visualization")
plt.xlabel("Sample mean x̄")
plt.ylabel("Density")
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# ============================
# PART (b): POWER AT μ = 3.25
# ============================

# Under alternative mu1_partb, x̄ ~ N(mu1_partb, sigma/sqrt(n))
# β = P(fail to reject H0 | μ = mu1_partb) = P(lower_crit < x̄ < upper_crit)
z_L = (lower_crit - mu1_partb) / se
z_U = (upper_crit - mu1_partb) / se
beta = norm.cdf(z_U) - norm.cdf(z_L)
power_partb = 1 - beta

print("=== Part (b): Power at μ = 3.25 ===")
print(f"z_L (using μ = 3.25): {z_L:.3f}")
print(f"z_U (using μ = 3.25): {z_U:.3f}")
print(f"β (Type II error): {beta:.4f}")
print(f"Power: {power_partb:.4f}")
print()

# Visualization of Part (b)
x_values_alt = np.linspace(mu0 - 4*se, mu1_partb + 4*se, 400)
pdf_H0 = norm.pdf(x_values_alt, loc=mu0, scale=se)
pdf_H1_b = norm.pdf(x_values_alt, loc=mu1_partb, scale=se)

plt.figure(figsize=(8, 5))
plt.plot(x_values_alt, pdf_H0, label="Under H₀: μ = 3.0")
plt.plot(x_values_alt, pdf_H1_b, linestyle="--", label="Under H₁: μ = 3.25")

# Shade power region under H1 (where we reject H0)
x_power_left = np.linspace(mu0 - 4*se, lower_crit, 200)
x_power_right = np.linspace(upper_crit, mu1_partb + 4*se, 200)
plt.fill_between(x_power_left,
                 norm.pdf(x_power_left, mu1_partb, se),
                 alpha=0.3,
                 label="Power region (left tail)")
plt.fill_between(x_power_right,
                 norm.pdf(x_power_right, mu1_partb, se),
                 alpha=0.3,
                 label="Power region (right tail)")

plt.axvline(lower_crit, linestyle=":", linewidth=1.5, label=f"Lower critical mean = {lower_crit:.2f}")
plt.axvline(upper_crit, linestyle=":", linewidth=1.5, label=f"Upper critical mean = {upper_crit:.2f}")

plt.title("Part (b): Power Visualization at μ = 3.25")
plt.xlabel("Sample mean x̄")
plt.ylabel("Density")
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# ============================
# PART (c): REQUIRED SAMPLE SIZE FOR POWER 0.9 AT μ = 3.75
# ============================

# Using formula:
# n = [ (z_{α/2} + z_{β}) * σ / (μ1 - μ0) ]^2
z_alpha_over_2 = norm.ppf(1 - alpha/2)
beta_target = 1 - power_target
z_beta_target = norm.ppf(1 - beta_target)

delta = abs(mu1_partc - mu0)

n_required_exact = ((z_alpha_over_2 + z_beta_target) * sigma / delta) ** 2
n_required = int(np.ceil(n_required_exact))  # round up

print("=== Part (c): Sample Size for Power 0.9 at μ = 3.75 ===")
print(f"Target power: {power_target}")
print(f"z_(α/2): {z_alpha_over_2:.3f}")
print(f"z_β (for β = {beta_target:.2f}): {z_beta_target:.3f}")
print(f"Exact n (not rounded): {n_required_exact:.3f}")
print(f"Required sample size (rounded up): n = {n_required}")
print()

# Verify power as a function of n around the required value
n_values = np.arange(5, 31)  # range of sample sizes to visualize
powers_c = []

for n_val in n_values:
    se_n = sigma / np.sqrt(n_val)
    # Critical means for this n
    lower_crit_n = mu0 - z_alpha_over_2 * se_n
    upper_crit_n = mu0 + z_alpha_over_2 * se_n
    # Under μ = mu1_partc:
    z_L_n = (lower_crit_n - mu1_partc) / se_n
    z_U_n = (upper_crit_n - mu1_partc) / se_n
    beta_n = norm.cdf(z_U_n) - norm.cdf(z_L_n)
    power_n = 1 - beta_n
    powers_c.append(power_n)

powers_c = np.array(powers_c)

# Visualization of Part (c): Power vs Sample Size
plt.figure(figsize=(8, 5))
plt.plot(n_values, powers_c, marker="o")
plt.axhline(power_target, linestyle="--", label=f"Target power = {power_target}")
plt.axvline(n_required, linestyle=":", label=f"Chosen n = {n_required}")

plt.title("Part (c): Power vs Sample Size at μ = 3.75")
plt.xlabel("Sample size n")
plt.ylabel("Power")
plt.grid(True, alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()
```

0.1.2 Two machines are used for filling plastic bottles with a net volume of 16.0 ounces. The fill volume can be assumed to be normal with standard deviation σ1 = 0.020 and σ2 = 0.025 ounces. A member of the quality engineering staff suspects that both machines fill to the same mean net volume, whether or not this volume is 16.0 ounces. A random sample of 10 bottles is taken from the output of each machine.

Machine 1 Machine 2 16.03 16.01 16.02 16.03 16.04 15.96 15.97 16.04 16.05 15.98 15.96 16.02 16.05 16.02 16.01 16.01 16.02 15.99 15.99 16.00

a. Do you think the engineer is correct? Use α = 0.05. What is the P-value for this test?

b. Calculate a 95% confidence interval on the difference in means. Provide a practical interpretation of this interval.

c. What is the power of the test in part (a) for a true difference in means of 0.04? d. Assume that sample sizes are equal. What sample size should be used to ensure that β = 0.05 if the true difference in means is 0.04? Assume that α = 0.05

Answer:


#### Given Data

| Parameter | Machine 1 (μ₁) | Machine 2 (μ₂) |
|-----------|----------------|----------------|
| Sample Size (n) | n₁ = 10 | n₂ = 10 |
| Pop. St. Dev. (σ) | σ₁ = 0.020 oz | σ₂ = 0.025 oz |
| **Significance Level (α):** | 0.05 | |

#### 1. Calculate Sample Means (x̄):

Based on the provided data table (10 samples per machine):

$$\bar{x}_1 = \frac{16.03 + 16.01 + 16.04 + \dots + 15.99}{10} = \mathbf{16.015} \text{ oz}$$

$$\bar{x}_2 = \frac{16.02 + 16.03 + 15.97 + \dots + 16.00}{10} = \mathbf{16.005} \text{ oz}$$

**Difference** (x̄₁ - x̄₂): 16.015 - 16.005 = **0.010 oz**

#### 2. Calculate Standard Error of the Difference (SE):

$$SE = \sqrt{\frac{\sigma_1^2}{n_1} + \frac{\sigma_2^2}{n_2}}$$

$$SE = \sqrt{\frac{0.020^2}{10} + \frac{0.025^2}{10}} = \sqrt{0.0001025} \approx \mathbf{0.010124}$$



#### Part a: Hypothesis Test and P-value (α = 0.05)

#### 1. Hypotheses:

* **Null Hypothesis (H₀):** μ₁ - μ₂ = 0 (The mean fill volumes are the same.)
* **Alternative Hypothesis (H₁):** μ₁ - μ₂ ≠ 0 (The mean fill volumes are different.)

#### 2. Test Statistic (Z-score):

$$Z = \frac{(\bar{x}_1 - \bar{x}_2) - (\mu_1 - \mu_2)_0}{SE}$$

$$Z = \frac{0.010 - 0}{0.010124} \approx \mathbf{0.988}$$

#### 3. P-value:

The P-value is the probability of observing a sample difference as extreme as 0.010 if H₀ were true. This is a two-tailed test:

$$P\text{-value} = 2 \times P(Z > |0.988|)$$

$$P(Z > 0.988) \approx 1 - 0.8385 = 0.1615$$

$$P\text{-value} = 2 \times 0.1615 \approx \mathbf{0.3230}$$

#### 4. Conclusion:

Since the P-value (**0.3230**) is much greater than the significance level (α = 0.05), we **Do Not Reject H₀**.






#### Part b: 95% Confidence Interval on the Difference in Means

The 95% CI is calculated using the formula: (x̄₁ - x̄₂) ± Z_(α/2) × SE

#### 1. Margin of Error (ME):

For a 95% CI, the critical Z-value is Z_(0.025) = **1.96**.

$$ME = 1.96 \times 0.010124 \approx \mathbf{0.01984}$$

#### 2. Confidence Interval:

$$\text{CI} = 0.010 \pm 0.01984$$

$$\text{CI} = [\mathbf{-0.00984}, \mathbf{0.02984}]$$


We are 95% confident that the true difference between the mean fill volume of Machine 1 and Machine 2 (μ₁ - μ₂) is between -0.00984 ounces and 0.02984 ounces. Since this interval includes zero, it suggests that a true difference of zero (i.e., the machines fill to the same mean volume) is a plausible value.



#### Part c: Power of the Test for a True Difference of 0.04

Power is the probability of correctly rejecting H₀ when the true difference in means (δ) is 0.04.

#### 1. Critical Differences (x̄_diff, crit):

The critical differences defining the rejection region under H₀ are:

$$\bar{x}_{\text{diff, crit}} = 0 \pm 1.96 \times 0.010124 \approx \pm \mathbf{0.01984}$$

**Rejection regions:** x̄_diff < -0.01984 or x̄_diff > 0.01984.

#### 2. Z-scores under the True Distribution (δ = 0.04):

We convert the critical differences to Z-scores using the true mean difference δ = 0.04:

$$Z_L = \frac{-0.01984 - 0.04}{0.010124} \approx \mathbf{-5.91}$$

$$Z_U = \frac{0.01984 - 0.04}{0.010124} \approx \mathbf{-1.99}$$

#### 3. Power Calculation:

$$\text{Power} = P(Z < Z_L) + P(Z > Z_U)$$

$$\text{Power} = P(Z < -5.91) + P(Z > -1.99)$$

$$P(Z < -5.91) \approx 0.0000$$

$$P(Z > -1.99) = 1 - P(Z < -1.99) \approx 1 - 0.0233 = 0.9767$$

$$\text{Power} \approx \mathbf{0.9767}$$

The power of the test to detect a true difference of 0.04 ounces is approximately 97.7%.



#### Part d: Required Sample Size (β = 0.05 for δ = 0.04)

We need to find n such that the power is 1 - β = 1 - 0.05 = 0.95. We assume n₁ = n₂ = n.

#### 1. Define Z-values:

* Z_(α/2) (α=0.05): **1.96**
* Z_β (for Power 0.95): **1.645**
* True Difference (δ): **0.04**

#### 2. Sample Size Formula (for equal n):

$$n = \frac{(Z_{\alpha/2} + Z_{\beta})^2 (\sigma_1^2 + \sigma_2^2)}{(\mu_1 - \mu_2)^2}$$

#### 3. Calculation:

$$n = \frac{(1.96 + 1.645)^2 (0.020^2 + 0.025^2)}{(0.04)^2}$$

$$n = \frac{(3.605)^2 (0.001025)}{0.0016} \approx 8.328$$

#### 4. Final Sample Size:

Since the sample size must be a whole number and we must ensure the power requirement is met, we round up:

$$n = \mathbf{9}$$

A sample size of n = **9** bottles from each machine (18 total) is required to ensure a Type II error probability (β) of no more than 0.05 when the true difference is 0.04 ounces.


```python
import numpy as np
from scipy.stats import norm
from math import ceil

# --- Data and Known Parameters ---

# Sample data provided in the problem
sample1 = np.array([16.03, 16.01, 16.04, 15.96, 16.05, 15.98, 16.05, 16.02, 16.02, 15.99])
sample2 = np.array([16.02, 16.03, 15.97, 16.04, 15.96, 16.02, 16.01, 16.02, 15.99, 16.00])

sigma1 = 0.020      # Population standard deviation for Machine 1
sigma2 = 0.025      # Population standard deviation for Machine 2
n1 = len(sample1)   # Sample size for Machine 1 (10)
n2 = len(sample2)   # Sample size for Machine 2 (10)
alpha = 0.05        # Significance level

# --- Step 1: Pre-calculations ---

x_bar1 = np.mean(sample1)
x_bar2 = np.mean(sample2)
x_bar_diff = x_bar1 - x_bar2

# Standard Error of the difference (SE)
SE_diff = np.sqrt((sigma1**2 / n1) + (sigma2**2 / n2))

print("--- Data Summary ---")
print(f"Machine 1 Mean (x̄₁): {x_bar1:.4f} oz")
print(f"Machine 2 Mean (x̄₂): {x_bar2:.4f} oz")
print(f"Observed Difference (x̄₁ - x̄₂): {x_bar_diff:.4f} oz")
print(f"Standard Error (SE_diff): {SE_diff:.4f}\n")


# ----------------------------------------------------------------------
# --- Part A: Hypothesis Test (H₀: μ₁ - μ₂ = 0) ---
# ----------------------------------------------------------------------

print("--- Part A: Hypothesis Test and P-value ---")

# Z-test statistic
Z_calc = x_bar_diff / SE_diff
p_value = 2 * (1 - norm.cdf(abs(Z_calc)))

print(f"1. Test Statistic (Z-score): {Z_calc:.4f}")
print(f"2. P-value (Two-tailed): {p_value:.4f}")

if p_value < alpha:
    decision = "Reject H₀"
else:
    decision = "Do Not Reject H₀"

print(f"3. Conclusion (α={alpha}): {decision}")
print("   Interpretation: The engineer's suspicion (that means are the same) cannot be refuted.\n")


# ----------------------------------------------------------------------
# --- Part B: 95% Confidence Interval (CI) ---
# ----------------------------------------------------------------------

print("--- Part B: 95% Confidence Interval ---")

Z_critical = norm.ppf(1 - alpha / 2) # 1.96 for 95% CI
ME = Z_critical * SE_diff            # Margin of Error

CI_L = x_bar_diff - ME
CI_U = x_bar_diff + ME

print(f"1. Margin of Error (ME): {ME:.4f}")
print(f"2. 95% CI for (μ₁ - μ₂): [{CI_L:.4f}, {CI_U:.4f}]")
print("   Practical Interpretation: Since the interval includes 0, it is plausible that the true mean difference is zero.\n")


# ----------------------------------------------------------------------
# --- Part C: Power Calculation (True Difference, δ = 0.04) ---
# ----------------------------------------------------------------------

print("--- Part C: Power Calculation ---")
delta_H1 = 0.04  # True difference in means (μ₁ - μ₂) under H₁

# 1. Critical Differences (x_crit) under H₀
x_crit_U = 0 + Z_critical * SE_diff
x_crit_L = 0 - Z_critical * SE_diff

# 2. Convert critical values to Z-scores under H₁ (mean = 0.04)
Z_L_under_H1 = (x_crit_L - delta_H1) / SE_diff
Z_U_under_H1 = (x_crit_U - delta_H1) / SE_diff

# 3. Power is P(Z < Z_L_under_H1) + P(Z > Z_U_under_H1)
power = norm.cdf(Z_L_under_H1) + (1 - norm.cdf(Z_U_under_H1))

print(f"1. Z-scores under H₁: Z_L={Z_L_under_H1:.2f}, Z_U={Z_U_under_H1:.2f}")
print(f"2. Power (1-β) for δ=0.04: {power:.4f} (or {power*100:.1f}%)")


# ----------------------------------------------------------------------
# --- Part D: Sample Size Calculation (β=0.05, α=0.05, δ=0.04) ---
# ----------------------------------------------------------------------

print("\n--- Part D: Required Sample Size ---")

beta = 0.05
Z_alpha_half = Z_critical # 1.96
Z_beta = norm.ppf(1 - beta) # 1.645 for 95% power (1-β)

# Sum of variances
sum_of_variances = sigma1**2 + sigma2**2

# Sample size formula for n₁=n₂=n (two-tailed test)
# n = [(Z_alpha_half + Z_beta)^2 * (σ₁² + σ₂²)] / (δ)²
n_required = ((Z_alpha_half + Z_beta)**2 * sum_of_variances) / (delta_H1**2)

# Round up to the next integer
n_final = ceil(n_required)

print(f"1. Theoretical required n (per machine): {n_required:.3f}")
print(f"2. Final required sample size (n): {n_final}")
print("   Interpretation: We need 9 bottles from each machine (18 total).\n")
```

#### Problem 11.2.1:

Diabetes and obesity are serious health concerns in the United States and much of the developed world. Measuring the amount of body fat a person carries is one way to monitor weight control progress, but measuring it accurately involves either expensive X-ray equipment or a pool in which to dunk the subject. Instead body mass index (BMI) is often used as a proxy for body fat because it is easy to measure:

$$\text{BMI} = \frac{\text{mass(kg)}}{\text{height(m)}^2} = 703 \frac{\text{mass(lb)}}{\text{height(in)}^2}$$

In a study of 250 men at Brigham Young University, both BMI and body fat were measured. Researchers found the following summary statistics:

$$\sum_{i=1}^{n} x_i = 6322.28$$

$$\sum_{i=1}^{n} x_i^2 = 162674.18$$

$$\sum_{i=1}^{n} y_i = 4757.90$$

$$\sum_{i=1}^{n} y_i^2 = 107679.27$$

$$\sum_{i=1}^{n} x_i y_i = 125471.10$$

a. Calculate the least squares estimates of the slope and intercept. Graph the regression line.

b. Use the equation of the fitted line to predict what body fat would be observed, on average, for a man with a BMI of 30.

c. Suppose that the observed body fat of a man with a BMI of 25 is 25%. Find the residual for that observation.

d. Was the prediction for the BMI of 25 in part (c) an overestimate or underestimate? Explain briefly.

Answer:


 n=250 men. The variables are BMI (x) and Body Fat (y).

#### Given Data

* n = 250
* $\sum x_i = 6322.28$
* $\sum x_i^2 = 162674.18$
* $\sum y_i = 4757.90$
* $\sum x_i y_i = 125471.10$


#### 1. Calculate Sample Means (x̄, ȳ):

* $\bar{x} = \frac{\sum x_i}{n} = \frac{6322.28}{250} = \mathbf{25.28912}$

* $\bar{y} = \frac{\sum y_i}{n} = \frac{4757.90}{250} = \mathbf{19.0316}$

#### 2. Calculate S_xx and S_xy:

* $S_{xx} = \sum x_i^2 - n\bar{x}^2 = 162674.18 - 250 \times (25.28912)^2 \approx \mathbf{2787.9875}$

* $S_{xy} = \sum x_i y_i - n\bar{x}\bar{y} = 125471.10 - 250 \times 25.28912 \times 19.0316 \approx \mathbf{5169.4975}$



#### Part a: Calculate the least squares estimates of the slope and intercept

#### 1. Calculate the Slope (β̂₁):

$$\hat{\beta}_1 = \frac{S_{xy}}{S_{xx}} = \frac{5169.4975}{2787.9875} \approx \mathbf{1.85434}$$

#### 2. Calculate the Intercept (β̂₀):

$$\hat{\beta}_0 = \bar{y} - \hat{\beta}_1 \bar{x} = 19.0316 - 1.85434 \times 25.28912 \approx \mathbf{-27.8724}$$

#### 3. Regression Line Equation:

The least squares regression line is $\hat{y} = \hat{\beta}_0 + \hat{\beta}_1 x$.

$$\hat{y} = \mathbf{-27.8724 + 1.8543 x}$$

The graph of this line would show a positive linear relationship, where body fat (y) increases by approximately 1.85 percentage points for every 1 unit increase in BMI (x).



#### Part b: Use the equation of the fitted line to predict what body fat would be observed, on average, for a man with a BMI of 30

To predict the body fat for x = 30:

$$\hat{y} = -27.8724 + 1.8543 (30)$$

$$\hat{y} = -27.8724 + 55.629$$

$$\hat{y} = \mathbf{27.7566\%}$$

The predicted average body fat for a man with a BMI of 30 is 27.76%.



#### Part c: Suppose that the observed body fat of a man with a BMI of 25 is 25%. Find the residual for that observation

#### 1. Predict Body Fat (ŷ) for x=25:

$$\hat{y} = -27.8724 + 1.8543 (25)$$

$$\hat{y} = -27.8724 + 46.3575$$

$$\hat{y} = 18.4851\%$$

#### 2. Calculate the Residual (e):

The residual is the difference between the observed value (y) and the predicted value (ŷ).

$$e = y - \hat{y}$$

$$e = 25 - 18.4851$$

$$e = \mathbf{6.5149}$$

The residual for the observation (BMI 25, Body Fat 25%) is 6.5149 percentage points.



#### Part d: Was the prediction for the BMI of 25 in part (c) an overestimate or underestimate? Explain briefly

The prediction was an **underestimate**.

**Explanation:** The residual is positive (e = 6.5149 > 0). A positive residual means the observed body fat (y = 25%) was higher than the predicted body fat (ŷ ≈ 18.49%). Therefore, the regression line underestimated the actual body fat for that individual.

```python
import numpy as np
import matplotlib.pyplot as plt

# --- Given Summary Statistics ---
# n = 250 men
# x = BMI
# y = Body Fat (%)
n = 250
sum_x = 6322.28
sum_x_sq = 162674.18
sum_y = 4757.90
sum_y_sq = 107679.27 # Note: Not strictly needed for a, b, c, d but good practice
sum_xy = 125471.10

# --- Preliminary Calculations ---

# 1. Calculate Sample Means (x_bar, y_bar)
x_bar = sum_x / n
y_bar = sum_y / n

# 2. Calculate S_xx and S_xy (required for slope)
S_xx = sum_x_sq - n * (x_bar**2)
S_xy = sum_xy - n * x_bar * y_bar

print("--- Preliminary Results ---")
print(f"Sample Mean BMI (x̄): {x_bar:.5f}")
print(f"Sample Mean Body Fat (ȳ): {y_bar:.5f}")
print(f"S_xx: {S_xx:.4f}")
print(f"S_xy: {S_xy:.4f}\n")

# ----------------------------------------------------------------------
# --- Part a: Calculate the least squares estimates of the slope and intercept ---
# ----------------------------------------------------------------------

# 1. Calculate the Slope (β̂₁)
beta_1 = S_xy / S_xx

# 2. Calculate the Intercept (β̂₀)
beta_0 = y_bar - beta_1 * x_bar

print("--- Part a: Regression Line ---")
print(f"Slope (β̂₁): {beta_1:.5f}")
print(f"Intercept (β̂₀): {beta_0:.4f}")
print(f"Fitted Regression Line: ŷ = {beta_0:.4f} + {beta_1:.4f}x\n")

# ----------------------------------------------------------------------
# --- Part b: Predict body fat for a man with a BMI of 30 ---
# ----------------------------------------------------------------------

bmi_b = 30
y_hat_b = beta_0 + beta_1 * bmi_b

print("--- Part b: Prediction for BMI = 30 ---")
print(f"Predicted Body Fat (ŷ at x=30): {y_hat_b:.4f}%")
print("This result indicates that, on average, a man with a BMI of 30 is predicted to have about 27.76% body fat.\n")

# ----------------------------------------------------------------------
# --- Part c & d: Residual for observation (BMI=25, Body Fat=25%) ---
# ----------------------------------------------------------------------

bmi_c = 25
y_observed_c = 25

# 1. Predict Body Fat (ŷ) for x=25
y_hat_c = beta_0 + beta_1 * bmi_c

# 2. Calculate the Residual (e)
residual_c = y_observed_c - y_hat_c

print("--- Part c & d: Residual Analysis for BMI = 25 ---")
print(f"Predicted Body Fat (ŷ at x=25): {y_hat_c:.4f}%")
print(f"Observed Body Fat (y at x=25): {y_observed_c:.2f}%")
print(f"Residual (e = y - ŷ): {residual_c:.4f}")

# Part d: Over/underestimate explanation
if residual_c > 0:
    d_explanation = "Underestimate (Observed > Predicted)"
else:
    d_explanation = "Overestimate (Observed < Predicted)"

print(f"\n--- Part d: Over/Underestimate ---")
print(f"The prediction was an: {d_explanation}")
print("Explanation: Since the residual is positive, the fitted line predicted a value lower than the observed value.\n")


# ----------------------------------------------------------------------
# --- Visualization (Illustrates Part a, b, c) ---
# ----------------------------------------------------------------------

# NOTE: We only have summary stats, so we cannot plot the actual 250 points.
# We will plot the regression line and key calculated points.

plt.figure(figsize=(8, 6))

# Plot the regression line (using a range around the sample mean)
x_plot = np.linspace(x_bar - 5, x_bar + 5, 100)
y_plot = beta_0 + beta_1 * x_plot
plt.plot(x_plot, y_plot, color='blue', label=r'Fitted Line: $\hat{y} = -27.8724 + 1.8543x$')

# Plot the mean point (x̄, ȳ) - the regression line must pass through this point
plt.plot(x_bar, y_bar, 'o', color='black', label=f'Sample Mean ({x_bar:.1f}, {y_bar:.1f})')

# Plot prediction for Part b (BMI = 30)
plt.plot(bmi_b, y_hat_b, 's', color='green', markersize=8, label=f'Prediction at BMI 30 ({y_hat_b:.1f}%)')

# Plot the observation and residual for Part c
plt.plot(bmi_c, y_observed_c, 'x', color='red', markersize=10, linewidth=2, label=f'Observed (25, 25%)')
plt.plot(bmi_c, y_hat_c, '^', color='red', markersize=8, label=f'Predicted (25, {y_hat_c:.1f}%)')
plt.vlines(bmi_c, y_hat_c, y_observed_c, color='red', linestyle=':', label=f'Residual ({residual_c:.2f})')

plt.title('Linear Regression of Body Fat vs. BMI')
plt.xlabel('BMI ($x$)')
plt.ylabel('Body Fat Percentage ($y$)')
plt.legend()
plt.grid(True, linestyle='--')
plt.show()
```
