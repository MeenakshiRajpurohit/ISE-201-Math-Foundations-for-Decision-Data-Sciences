```python
import numpy as np
import matplotlib.pyplot as plt

# --- 1. Define the CDF Function ---
def cdf_pareto(x):
    """CDF for F(x) = 1 - 1/x^2, defined for x >= 1"""
    # Use np.where to ensure the function is 0 for x < 1
    return np.where(x >= 1, 1 - 1 / x**2, 0)

# --- 2. Calculate the Probability ---
x_val = 2
probability = cdf_pareto(x_val) # P(X < 2) = F(2)

# --- 3. Generate Plot Data ---
x = np.linspace(0.5, 5.0, 500)
y = cdf_pareto(x)

# --- 4. Create the Visualization ---
plt.figure(figsize=(8, 5))

# Plot the CDF curve
plt.plot(x, y, label=r'$F(x) = P(X \leq x)$', color='blue')

# Highlight the specific calculation point (x=2)
plt.vlines(x_val, 0, probability, color='red', linestyle='--', linewidth=1, label=r'$x = 2$')
plt.hlines(probability, 0.5, x_val, color='red', linestyle='--', linewidth=1)

# Mark the calculated point
plt.plot(x_val, probability, 'ro', markersize=6)

# Annotate the result
plt.annotate(
    f'$P(X < 2) = {probability:.2f}$',
    xy=(x_val, probability),
    xytext=(2.5, 0.65),
    arrowprops=dict(facecolor='black', shrink=0.05, width=0.5, headwidth=6),
    fontsize=12,
    color='darkgreen'
)
plt.annotate(
    r'$x=1$ (Start)',
    xy=(1, 0),
    xytext=(1.05, 0.1),
    fontsize=10,
    color='black'
)

# Set plot labels and title
plt.title('Cumulative Distribution Function (CDF) for $F(x) = 1 - 1/x^2$')
plt.xlabel('X (Random Variable Value)')
plt.ylabel(r'$F(x) = P(X \leq x)$')
plt.xlim(0.5, 5)
plt.ylim(0, 1.05)
plt.grid(True, linestyle=':', alpha=0.6)
plt.legend(loc='lower right')
plt.show()
```
