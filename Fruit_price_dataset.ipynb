```python
# --------------------------------------------------------
# Descriptive Statistics for Fruit Prices Dataset (2022)
# --------------------------------------------------------

# 1 Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 2 Load the dataset
# Make sure the file path is correct
file_path = "/content/Fruit-Prices-2022.csv"
data = pd.read_csv(file_path)

# 3 Display the first few rows of the dataset
print("=== Preview of the Dataset ===")
print(data.head(), "\n")

# 4 Check basic information about the dataset
print("=== Dataset Information ===")
print(data.info(), "\n")

# 5 Check for missing values
print("=== Missing Values in Each Column ===")
print(data.isnull().sum(), "\n")

# 6 Get column-wise data types
print("=== Data Types ===")
print(data.dtypes, "\n")

# 7 Generate descriptive statistics for numerical columns
print("=== Descriptive Statistics (Numerical Columns) ===")
print(data.describe(), "\n")

# 8 Generate descriptive statistics for categorical columns
print("=== Descriptive Statistics (Categorical Columns) ===")
print(data.describe(include=['object']), "\n")

# 9 Get number of unique values for each column
print("=== Unique Values in Each Column ===")
for col in data.columns:
    print(f"{col}: {data[col].nunique()} unique values")

print("\n")

#  View correlations between numeric variables (if any)
print("=== Correlation Matrix ===")
print(data.corr(numeric_only=True), "\n")

# 11 Visualize correlations using a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(data.corr(numeric_only=True), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap - Fruit Prices Data")
plt.show()

# 12 Visualize distributions of numeric columns
numeric_cols = data.select_dtypes(include=[np.number]).columns
for col in numeric_cols:
    plt.figure(figsize=(7, 4))
    sns.histplot(data[col], kde=True, bins=15)
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()

# 13 Example: Average price per fruit
if "Fruit" in data.columns and "RetailPrice" in data.columns:
    avg_price = data.groupby("Fruit")["RetailPrice"].mean().sort_values(ascending=False)
    print("=== Average Retail Price per Fruit ===")
    print(avg_price, "\n")

    plt.figure(figsize=(10, 6))
    avg_price.plot(kind="bar", color="skyblue")
    plt.title("Average Retail Price per Fruit")
    plt.xlabel("Fruit")
    plt.ylabel("Average Price ($)")
    plt.xticks(rotation=90)
    plt.tight_layout()
    plt.show()

# 14 Example: Price per cup equivalent (if columns exist)
if "CupEquivalentPrice" in data.columns:
    plt.figure(figsize=(10, 6))
    sns.boxplot(x="Form", y="CupEquivalentPrice", data=data)
    plt.title("Cup Equivalent Price by Form")
    plt.xlabel("Form")
    plt.ylabel("Cup Equivalent Price ($)")
    plt.tight_layout()
    plt.show()

# 15 Summary
print(" Descriptive statistics and visualizations generated successfully.")
```
